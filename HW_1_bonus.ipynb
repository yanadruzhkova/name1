```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Mountains.csv')
data.head()
```

```python
data1 = pd.pivot_table(data, index = 'Parent mountain', values = 'Mountain', aggfunc ='count' )
```

```python
data2 = data1.loc[data1['Mountain'] >3]
```

```python
data3 = data2.sort_values( by = 'Mountain', ascending = False ).reset_index() 
data3
```

## Бонусная задача

```python
data4 = data.merge( data3, how = 'left' , left_on = 'Parent mountain', right_on = 'Parent mountain')
```

```python
data5 = data4.loc[data4['Mountain_y'] >3]
```

```python
final_data = data5.drop('Mountain_y', axis=1)
final_data
```
